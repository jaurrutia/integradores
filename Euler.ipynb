```python
def euler(funcion,p_ini,t_final,h=0.01):  #p_ini contiene toda la info de la función---> [t_0,f_0x,f_0y,f_0z,etc]
    tiempos = arange(p_ini[0],t_final+h,h)     #variable
    y= [0]*(len(tiempos))
    y[0]=p_ini[1]
    for i in range(len(tiempos)-1):
        y[i+1] = y[i]+funcion(y[i],tiempos[i])*h
    solucion=[tiempos,y]
    return solucion
```

```python
def funcion_bonita(y,t):
    return ((y*2./sqrt(3.))**2-1)*(1./(sin(t)+2))
```

```python
hs= [0.1,0.01,0.001,.00001]
for h in hs:
    sol=euler(funcion_bonita,[1,1],3.,h)
    plot(sol[0],sol[1], label="h %h")
    legend()
show()
```
